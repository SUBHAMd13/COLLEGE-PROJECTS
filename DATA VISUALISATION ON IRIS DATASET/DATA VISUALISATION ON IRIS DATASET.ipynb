```python
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    iris = pd.read_csv("/home/subham/Downloads/DATA VISUALISATION ON IRIS DATASET/Iris.csv")
    print (iris.shape)
    
    
```

In this dataset contains 4 types species(label).These are Iris-setosa, Iris-versicolor, Iris-verginica.

# 2-D SCATTER PLOT

```python
iris.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm')
plt.show()
```

Here I use scatter-plot in features SepalLength and SepalWidth. But this plot is not so useful to distinguise.

```python
sns.FacetGrid(iris, hue="Species", size=5).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
plt.show()
```

Now when I use some color to distinguise them then Iris-setosa fully separated from other two species.

# PAIR-PLOT

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
plt.show()
```

Here I use pairplot to visualise some distinguisable pattern using all 4 features.
In this plot using PetalLength and PetalWidth I can patially differentiate Iris-versicolor and Iris-verginica.

# PDF OF PETAL LENGTH

```python
sns.FacetGrid(iris.drop("Id", axis=1), hue="Species",size=6).map(sns.kdeplot,"PetalLengthCm").add_legend()
plt.show()
```

Here I use KDEplot to pdf plotting to watching overlapping situation of versicolor and verginica.

# BOX WHISHKERS PLOT

```python
sns.boxplot(x='Species',y='PetalLengthCm', data=iris)
plt.show()
```

Using Box and Whiskers plot we can observe the spacings between the different parts of the box 
indicate the degree of dispersion (spread) and skewness in the data, and show outliers.
There is 3 quertiles. First quartile lower quartile which is 25th percentile, 
second quartile median 50th percentile, third quartile upper quartile 75th percentile.
In this plot 3rd quertile of versicolor overlapped with the 1st quertile of virginica.

# CDF OF EACH FEATURES

```python
ax = sns.kdeplot(iris['SepalLengthCm'], cumulative=True)
ax = sns.kdeplot(iris['SepalWidthCm'], cumulative=True)
ax = sns.kdeplot(iris['PetalLengthCm'], cumulative=True)
ax = sns.kdeplot(iris['PetalWidthCm'], cumulative=True)
plt.show()
```

Here I use CDF using kdeplot the cdf gives the probability under a certain point.
Cumulative Distribution Function plots display exactly the same information as do histograms.
CDF plots are extremely useful for quickly finding the fluorescence of a distribution corresponding
to any given percentile (such as the median). 
This plot shows how to find the median (50th percentile), 75th, and 90th percentiles for these features in data.
