# PAIRPLOT

From above scatterplots we can't decide anything. Here also overlapping spread through out the space.
Another useful seaborn plot is the pairplot, which shows the bivariate relation between each pair of features.
So draw a pairplot on 3 features.

```python
sns.pairplot(haberman, hue="survived",vars=("age","year","axillary"), size=3, diag_kind="kde")
plt.show()
```

```python
sns.pairplot(haberman, hue="survived",vars=("age","year","axillary"), size=3, diag_kind="hist")
plt.show()
```

# PDF

Through above pairplot nothing can be determined.Draw the pdf of each features.
A final seaborn plot useful for looking at univariate relations is the kdeplot,

```python
sns.FacetGrid(haberman, hue="survived",size=6).map(sns.kdeplot,"axillary").add_legend()
plt.show()
```

# BOXPLOT 

Now we will try boxplot to visualize each features.
We can look at an individual feature in Seaborn through a boxplot.

```python
sns.boxplot(x='survived',y='axillary', data=haberman)
plt.show()
```

```python
sns.boxplot(x='survived',y='age', data=haberman)
plt.show()
```

```python
sns.boxplot(x='survived',y='year', data=haberman)
plt.show()
```

One way we can extend this plot is adding a layer of individual points on top of
it through Seaborn's striplot.
 
 We'll use jitter=True so that all the points don't fall in single vertical lines
 above the survived.

 Saving the resulting axes as ax each time causes the resulting plot to be shown
 on top of the previous axes

```python
ax = sns.boxplot(x='survived',y='axillary', data=haberman)
plt.show()
ax = sns.stripplot(x='survived',y='axillary', data=haberman, jitter=True, edgecolor="gray")
plt.show()
```

```python
ax = sns.boxplot(x='survived',y='age', data=haberman)
plt.show()
ax = sns.stripplot(x='survived',y='age', data=haberman, jitter=True, edgecolor="gray")
plt.show()
```

```python
ax = sns.boxplot(x='survived',y='year', data=haberman)
plt.show()
ax = sns.stripplot(x='survived',y='year', data=haberman, jitter=True, edgecolor="gray")
plt.show()
```

We can make a boxplot with Pandas on each feature split out by survived feature.

```python
haberman.boxplot(by="survived", figsize=(12, 6))
plt.show()
```

# CDF

the cumulative distribution function (CDF) of a real-valued random variable X, or just distribution function of X, evaluated at x, is the probability that X will take a value less than or equal to x.

CDF also used to test to see whether two empirical distributions are different or whether an empirical distribution is different from an ideal distribution.

Plot CDF of each feature and also cummulate them.

```python
ax = sns.kdeplot(haberman['axillary'], cumulative=True)

plt.show()
```

```python
ax = sns.kdeplot(haberman['age'], cumulative=True)

plt.show()
```

```python
ax = sns.kdeplot(haberman['year'], cumulative=True)

plt.show()
```

```python
ax = sns.kdeplot(haberman['axillary'], cumulative=True)
ax = sns.kdeplot(haberman['age'], cumulative=True)
ax = sns.kdeplot(haberman['year'], cumulative=True)
plt.show()
```

# PARALLEL CO-ORDINATES PLOT

Another multivariate visualization technique pandas has is parallel_coordinates.
Parallel coordinates plots each feature on a separate column & then draws lines connecting the features for each data sample.

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(haberman, "survived")
plt.show()
```

# ANDREWS CURVES

One cool more sophisticated technique pandas has available is called Andrews Curves.
Andrews Curves involve using attributes of samples as coefficients for Fourier series and then plotting these.

```python
from pandas.plotting import andrews_curves
andrews_curves(haberman, "survived")
plt.show()
```

# RADVIZ PLOT

Another multivariate visualization technique pandas has is radviz.
Which puts each feature as a point on a 2D plane, and then simulates having each sample attached to those points through a spring weighted by the relative value for that feature.

```python
from pandas.plotting import radviz
radviz(haberman, "survived")
plt.show()
```
